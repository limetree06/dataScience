## 결측치 다루기

```python
df.isnull() #False -> 값이 있다. #True -> 값이 없다.

# True 는 1이니깐 sum하면 결측치의 개수를 알수 있다. 결측치 확인하기
null_count = df.isnull().sum()
#bar + h(90도 돌려줌)
#figsize -> 그래프 사이즈
null_count.plot.barh(figsize=(5,7)) 
#plot하고 tab하면 python에서 쓸수있는 그래프 종류 뜬다

df_null_count = null_count.reset_index() #reset_index() : 데이터 프레임으로 변경해준다.

#컬럼명 변경
df_null_count.columns = ["컬럼명", "결측치 수"]
df_null_count.head()
```

```python
#df_null_count 데이터 프레임에 있는 결측치 수를 sort_values를 통해 정렬해서
# 결측치가 많은 순으로 상위 10개만 출력한다.
#결측치가 너무 커서 분석에 방해되면 결측치를 제거해주는 것도 방법이 된다.
df_null_count_top = df_null_count.sort_values(by="결측치 수", ascending=False).head(10)

#NaN == Not a Number의 약자
#to_list() 리스트 형태로 변경해준다.
drop_columns = df_null_count_top["컬럼명"].to_list()

#axis = 0 행을 기준으로 drop
#axis 0,1순 / 행렬 순
df = df.drop(columns=drop_columns, axis = 1)
df.shape
df.info()
```

## 수치데이터 요약하기

```python
#데이터 타입이 int나 float처럼 수치형 데이터에 한해서만 기초 통계값을 구해볼수 있다.
df.dtypes
df["위도"].mean() #평균값
df["위도"].median() #중앙값 -> 값들의 중앙(최솟값과 최댓값에 영향을 많이 받지 않는다.)
df["위도"].max() #최댓값
df["위도"].min() #최솟값
df["위도"].count() #갯수 -> 결측치를 제외한 갯수
df["위도"].describe() #데이터를 요약해서 보여준다. 25%는 25%에 있는 값
#df["위도", "경도"] # 오류란다. 판다스에서 두개이상의 자료구조를 가져올때는 리스트 형태로 가져와야한다.
df[["위도","경도"]].describe() # 2개의 컬럼을 descrive로 요약

#문자 형태로만 되어있는 데이터타입이 요약이 된다.
# top 가잔많이 등장하는 문자열
# freq 빈도수 몇번 등장하는지를 나타낸다.
# 결측치는 빼고 보여진다.
df.describe(include = 'all')
df.describe(include = 'object')
```

## 데이터 색인하기
- 특정데이터만 모아서 따로 보는 것

```python
# "상권업종중분류명" 이 "약국/한약방"인 데이터만 가져온다.
#특정 값을 바꿀때 copy() 하지 않고 바꾸면 df값도 바뀌게 된다.

print()

df_medical = df[df["상권업종중분류명"] == "약국/한약방"].copy()

#loc -> 행과 열을 함께 가져올수 있다.

m = df["상권업종대분류명"] == '의료'
df.loc[df["상권업종대분류명"] == '의료']["상권업종중분류명"] # df에 두번 접근하는 구조다
df.loc[df["상권업종대분류명"]=="의료", "상권업종중분류명"].value_counts()
df.loc[m, "상권업종중분류명"].value_counts()

#의복의류 의 개수를 알수있다.
df_medi = df[df["상권업종중분류명"] =="유사의료업"]

#의복의류 회사 상위 10개
df_medi["상호명"].value_counts().head(10)

#여러 조건으로 색인하기
# series ambious error 판다스에서 and일때는 & 로 표시
df_seoul_drug = df[(df["상권업종소분류명"] == "약국") & (df["시도명"] == "서울특별시")].copy()
df_seoul_drug.shape
```

## 텍스트 데이터로 전처리하기
- 원하는 텍스트 데이터를 추출하고 처리하는 방법

```python
#value_counts() 개수가 많은 순으로 정렬해서 보여준다.
c = df_seoul_drug["시군구명"].value_counts(normalize=True)
#c.plot.bar(rot=60)

df_seoul_hospital = df[(df["상권업종소분류명"]=="종합병원") & (df["시도명"] =="서울특별시")].copy()
df_seoul_hospital["시군구명"].value_counts()

#텍스트 데이터 색인하기 -> str.contains("text") 특정 text가 포함된 애만 들고온다.
#커피라는 단어가 안들어간 애들만 가져온다.
#상호명만 보고 싶은때 .loc
df_seoul_hospital.loc[~df_seoul_hospital["상호명"].str.contains('종합병원'), "상호명"].unique()

#이런 애들이 있음 -> 종합병원이라고 보기 어렵다
df_seoul_hospital[df_seoul_hospital["상호명"].str.contains('꽃배달')]
df_seoul_hospital[df_seoul_hospital["상호명"].str.contains('의료기')]

drop_row = df_seoul_hospital[df_seoul_hospital["상호명"].str.contains('의료기|꽃배달|어린이집|상담소|장례식장')].index
drop_row = drop_row.tolist()

drop_row2 = df_seoul_hospital[df_seoul_hospital["상호명"].str.endswith("의원")].index
drop_row2 = drop_row2.tolist()
drop_row = drop_row + drop_row2

#axis=0 행을 기준으로 드랍
df_seoul_hospital = df_seoul_hospital.drop(drop_row, axis=0)

plt.figure(figsize=(15,4))
sns.countplot(data=df_seoul_hospital, x="시군구명",
                  order=df_seoul_hospital["시군구명"].value_counts().index)
```

## folium으로 위경도와 주소 데이터를 지도에 표현
- https://nbviewer.org/github/python-visualization/folium/blob/main/examples/Quickstart.ipynb

```python
import folium

lat_mean = df_seoul_hospital["위도"].mean()
long_mean = df_seoul_hospital["경도"].mean()

map = folium.Map(location = [lat_mean,long_mean], zoom_start=12)

for n in df_seoul_hospital.index:
    name = df_seoul_hospital.loc[n, "상호명"]
    address = df_seoul_hospital.loc[n, "도로명주소"]
    popup = f"{name}-{address}"
    location = [df_seoul_hospital.loc[n, "위도"], df_seoul_hospital.loc[n, "경도"]]
    folium.Marker(
        location = location, 
        popup = popup).add_to(map)
map
```
