```python
import requests
import re
import bs4

url = "https://core.today/story/explore/"
response = requests.get(url)
bs = bs4.BeautifulSoup(response.content, "html.parser")


#search_data = bs.find_all('div', class_ = "col-md-7 mt-3 mt-md-0")
#search_data = bs.select('div a')
search_data = bs.find_all('div', class_ = "col-md-7 mt-3 mt-md-0")

for content_tag in search_data:
    tag_name = content_tag.find('a', class_="badge bg-primary-soft text-primary mb-2")
    print( )
    print(tag_name)
```
